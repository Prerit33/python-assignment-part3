```python
#TASK1:

filename = "python_notes.txt"

# 1. The initial lines to write
initial_notes = [
    "Topic 1: Variables store data. Python is dynamically typed.\n",
    "Topic 2: Lists are ordered and mutable.\n",
    "Topic 3: Dictionaries store key-value pairs.\n",
    "Topic 4: Loops automate repetitive tasks.\n",
    "Topic 5: Exception handling prevents crashes.\n"
]

# Writing the lines to the file using write mode ('w') and UTF-8 encoding
with open(filename, 'w', encoding="utf-8") as file:
    file.writelines(initial_notes)
print("File written successfully.")

# 2. Appending two new lines of your own using append mode ('a')
additional_notes = [
    "Topic 6: Functions promote code reusability and organization.\n",
    "Topic 7: List comprehensions offer a concise way to create lists.\n"
]

with open(filename, 'a', encoding="utf-8") as file:
    file.writelines(additional_notes)
print("Lines appended successfully.\n")

# 3. Reading the file back, strip trailing newlines, and print numbered lines
lines_list = []
with open(filename, 'r', encoding="utf-8") as file:
    lines_list = file.readlines()

print("--- File Contents ---")
for index, line in enumerate(lines_list, start=1):
    # .strip() removes the trailing newline character and surrounding whitespace
    print(f"{index}. {line.strip()}")

# 4. Counting and printing the total number of lines
total_lines = len(lines_list)
print(f"\nTotal number of lines in the file: {total_lines}\n")

# 5. Asking the user for a keyword and perform a case-insensitive search
keyword = input("Enter a keyword to search for in the notes: ").strip()

# Converting the keyword to lowercase for a case-insensitive comparison
keyword_lower = keyword.lower()
matching_lines = []

for line in lines_list:
    if keyword_lower in line.lower():
        matching_lines.append(line.strip())

# Printing the results of the search
if matching_lines:
    print(f"\nFound {len(matching_lines)} line(s) containing '{keyword}':")
    for match in matching_lines:
        print(f" - {match}")
else:
    print(f"\nSorry, no lines were found containing the keyword '{keyword}'.")
```

```python
import requests
import json

# --- Step 1: Fetch and Display Products ---
print("--- Step 1: Fetch and Display Products ---\n")
url_20_products = "https://dummyjson.com/products?limit=20"
response = requests.get(url_20_products)
response.raise_for_status()  # Check for any HTTP errors
products = response.json().get('products', [])

# Printing the formatted table header
print(f"{'ID':<4} | {'Title':<30} | {'Category':<15} | {'Price':<8} | {'Rating'}")
print("-" * 75)

# Printing each product row
for p in products:
    # Truncate title and category just in case they exceed column widths
    title = str(p.get('title', ''))[:30]
    category = str(p.get('category', ''))[:15]
    price = p.get('price', 0)
    rating = p.get('rating', 0)
    print(f"{p.get('id', ''):<4} | {title:<30} | {category:<15} | ${price:<7.2f} | {rating}")


# --- Step 2: Filtering and Sorting ---
print("\n\n--- Step 2: Filter and Sort (Rating >= 4.5, Price Descending) ---\n")

filtered_products = [p for p in products if p.get('rating', 0) >= 4.5]

sorted_products = sorted(filtered_products, key=lambda x: x.get('price', 0), reverse=True)

print(f"{'ID':<4} | {'Title':<30} | {'Category':<15} | {'Price':<8} | {'Rating'}")
print("-" * 75)
for p in sorted_products:
    title = str(p.get('title', ''))[:30]
    category = str(p.get('category', ''))[:15]
    price = p.get('price', 0)
    rating = p.get('rating', 0)
    print(f"{p.get('id', ''):<4} | {title:<30} | {category:<15} | ${price:<7.2f} | {rating}")


# --- Step 3: Searching by Category ---
print("\n\n--- Step 3: Search by Category (Laptops) ---\n")
url_laptops = "https://dummyjson.com/products/category/laptops"
laptops_response = requests.get(url_laptops)
laptops_response.raise_for_status()
laptops = laptops_response.json().get('products', [])

for laptop in laptops:
    print(f"Title: {laptop.get('title')} | Price: ${laptop.get('price', 0):.2f}")


# --- Step 4: POST Request (Simulated) ---
print("\n\n--- Step 4: POST Request (Simulated) ---\n")
url_add = "https://dummyjson.com/products/add"
new_product_payload = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

# Send the POST request with the JSON payload
post_response = requests.post(url_add, json=new_product_payload)

# Printing the full response returned by the server
print("Server Response Code:", post_response.status_code)
print("Server Response JSON:")
print(json.dumps(post_response.json(), indent=2))
```

```python
#TASK3

import requests
import json

# Part A: Guarded Calculator
# ==========================================
print("--- Part A: Guarded Calculator ---")

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

# Testing Part A
print(f"10 / 2 = {safe_divide(10, 2)}")
print(f"10 / 0 = {safe_divide(10, 0)}")
print(f"'ten' / 2 = {safe_divide('ten', 2)}")


# Part B: Guarded File Reader
# ==========================================
print("\n--- Part B: Guarded File Reader ---")

def read_file_safe(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            return file.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return None
    finally:
        print("File operation attempt complete.")

# Testing Part B
print("\nTesting valid file (assuming python_notes.txt exists from Task 1):")
notes_content = read_file_safe("python_notes.txt")
if notes_content:
    print("-> Successfully read python_notes.txt!")

print("\nTesting invalid file:")
ghost_content = read_file_safe("ghost_file.txt")

# Part C: Robust API Calls
# ==========================================
print("\n--- Part C: Robust API Calls (Updated from Task 2) ---")

print("Attempting GET request...")
try:
    get_response = requests.get("https://dummyjson.com/products?limit=20", timeout=5)
    get_response.raise_for_status() # Check for HTTP errors
    print("-> GET request successful!")
except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"An unexpected error occurred during GET: {e}")



print("\nAttempting POST request...")
try:
    post_payload = {"title": "Robust Product", "price": 499, "category": "electronics"}
    post_response = requests.post("https://dummyjson.com/products/add", json=post_payload, timeout=5)
    post_response.raise_for_status()
    print("-> POST request successful! ID assigned:", post_response.json().get('id'))
except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"An unexpected error occurred during POST: {e}")



# Part D: Input Validation Loop
# ==========================================
print("\n--- Part D: Input Validation Loop ---")

while True:
    user_input = input("\nEnter a product ID to look up (1-100), or 'quit' to exit: ").strip()
    
    # CheckING for exit command
    if user_input.lower() == 'quit':
        print("Exiting the lookup tool. Goodbye!")
        break
    
    # Validate that input is an integer
    try:
        product_id = int(user_input)
    except ValueError:
        print("Warning: Invalid input. Please enter a whole number.")
        continue # Skip the rest of the loop and prompt again
    
    # Validate the range
    if not (1 <= product_id <= 100):
        print("Warning: Product ID out of bounds. Please enter a number between 1 and 100.")
        continue
    
    # MakING the API Call with robust error handling
    url = f"https://dummyjson.com/products/{product_id}"
    try:
        response = requests.get(url, timeout=5)
        
        # CheckING specific status codes
        if response.status_code == 200:
            product_data = response.json()
            title = product_data.get('title', 'Unknown Title')
            price = product_data.get('price', 0)
            print(f"Success! Found Product: {title} | Price: ${price}")
        elif response.status_code == 404:
            print("Product not found.")
        else:
            print(f"Unexpected HTTP status code received: {response.status_code}")
            
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"An unexpected system error occurred: {e}")
```

```python
#TASK4:

import requests
from datetime import datetime
import os

# Defining the log file name
LOG_FILE = "error_log.txt"


# 1. Error Logger Function
# ==========================================
def log_error(context, error_type, message):
    """Writes a timestamped error message to the log file."""
    # Get current time
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    # Constructing the log entry exactly as requested
    log_entry = f"[{timestamp}] ERROR in {context}: {error_type} — {message}\n"
    
    # Open the file in append mode ('a') with utf-8 encoding
    with open(LOG_FILE, 'a', encoding='utf-8') as file:
        file.write(log_entry)


# 2. Triggering Errors Intentionally
# ==========================================
print("Running simulated tasks to trigger errors...")

# Trigger 1: ConnectionError
print("-> Attempting to connect to an unreachable URL...")
try:
    # Intentionally bad URL to force a ConnectionError
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=3)
except requests.exceptions.ConnectionError:
    log_error("fetch_products", "ConnectionError", "No connection could be made")
    print("   Caught ConnectionError and logged it.")
except Exception as e:
    # Fallback for any other unexpected exception
    log_error("fetch_products", type(e).__name__, str(e))

# Trigger 2: HTTP 404 Error (Not an exception, handled via status_code)
print("-> Attempting to fetch a non-existent product ID (999)...")
try:
    response = requests.get("https://dummyjson.com/products/999", timeout=5)
    
    # Checking if the status code indicates an error (e.g., != 200)
    if response.status_code != 200:
        log_error(
            "lookup_product", 
            "HTTPError", 
            f"{response.status_code} Not Found for product ID 999"
        )
        print(f"   Caught HTTP {response.status_code} and logged it.")
        
except requests.exceptions.RequestException as e:
    # This catches network-level exceptions if the dummy API were completely down
    log_error("lookup_product", type(e).__name__, str(e))

# 3. Reading and Printing the Log File
# ==========================================
print("\n--- Contents of error_log.txt ---")
if os.path.exists(LOG_FILE):
    with open(LOG_FILE, 'r', encoding='utf-8') as file:
        print(file.read())
else:
    print("Error: The log file was not created.")
```
